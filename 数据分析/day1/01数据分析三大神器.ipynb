```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 创建数据
# 使用随机数创建五位同学三门课的成绩
import random
```

```python
scores = [[random.randint(60,100) for i in range(3)] for i in range(5)]
print(scores)
```

```python
# numpy ndarray 数据类型（处理多维数组）
```

```python
# 1.将列表转换为ndarray数据类型
scores = np.array(scores)
print(scores)
```

```python
#取最大值
scores.max()
```

```python
scores.max(axis=0)
```

```python
scores.max(axis=1)
```

```python
#最小值
scores.min(axis=1)
```

```python
#平均值
scores.mean(axis=1)
```

```python
#pandas
#DataFrame 数据类型（数据表）
name = ['吕布', '庄周', '狄仁杰', '刘备', '关羽']
courses = ['语文', '数学', '英语']
df = pd.DataFrame(data=scores, index=name, columns=courses)
```

```python
df
```

```python
# 每门课，每个人的平均成绩
df.mean(axis=1)
```

```python
np.round(df.mean(axis=1), 2)
```

```python
df
```

```python
df['平均分'] = np.round(df.mean(axis=1), 2)
```

```python
df
```

```python
df.loc['最高分'] = df.max()
```

```python
df.loc['最低分'] = df.min()
```
