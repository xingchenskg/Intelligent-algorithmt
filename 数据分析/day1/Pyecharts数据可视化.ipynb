```python
#调用类
from pyecharts import options as opts
#饼图调用的模块
from pyecharts.charts import Pie
#假数据
from pyecharts.faker import Faker
```

```python
#制作饼图
Faker.choose()
```

```python
Faker.values()
```

```python
a = zip(Faker.choose(), Faker.values())
```

```python
for i in a:
    print(i)
```

```python
c = (
    Pie()
    .add("图表一",[z for z in zip(Faker.choose(), Faker.values())])
    #全局配置项
    .set_global_opts(title_opts=opts.TitleOpts(title="图表练习", subtitle="我是副标题"))
    #系列配置项
    .set_series_opts(label_opts=opts.LabelOpts(formatter="{b}: {c}"))
    #生成图表
    .render_notebook()
)
```

```python
c
```

```python
#柱状图
from pyecharts.charts import Bar
```

```python
a = (
    Bar()
    .add_xaxis(Faker.choose())
#     .add_yaxis("商家A"，Faker.values())
    .add_yaxis("商家A", Faker.values())
    
    .set_global_opts(
        title_opts = opts.TitleOpts(title="主标题", subtitle="副标题"),
        yaxis_opts = opts.AxisOpts(name='Y轴'),
        xaxis_opts = opts.AxisOpts(name='X轴')
    )
    .render_notebook()
)
```

```python
a
```

```python
b = (
    Bar()
    .add_xaxis(Faker.choose())
#     .add_yaxis("商家A"，Faker.values())
    .add_yaxis("商家A", Faker.values())
    .add_yaxis("商家B", Faker.values())
    
    .set_global_opts(
        title_opts = opts.TitleOpts(title="主标题", subtitle="副标题"),
        yaxis_opts = opts.AxisOpts(name='Y轴'),
        xaxis_opts = opts.AxisOpts(name='X轴')
    )
    .render_notebook()
)
```
