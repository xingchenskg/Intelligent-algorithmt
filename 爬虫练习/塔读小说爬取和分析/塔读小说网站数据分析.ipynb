```python
# 导入数据
df=pd.read_excel('./data_house.xls',index_col=0)
df
```

```python
# 缺失值数量
df.isnull().sum()
```

```python
# 删除重复值, 去除完全重复的行数据
df.drop_duplicates(inplace=True)
df
```

```python
# 添加频道
df['channel']='男'
# df
```

```python
# df
```

```python
# 清洗数据
# 替换中文单位
# 替换words
def handle(words):
    if type(words) == str:
        if words[-1]=='万':
            words=float(words[:-1])*10000
        elif words[-1]=='千':
            words=float(words[:-1])*1000
        else:
            words =float(words)
    return words
df.words=df.words.map(lambda words:handle(words))
df
```

```python
# 替换weekpopularity
def handle(weekpopularity):
    if type(weekpopularity) == str:
        if weekpopularity[-1]=='万':
            weekpopularity=float(weekpopularity[:-1])*10000
        elif weekpopularity[-1]=='千':
            weekpopularity=float(weekpopularity[:-1])*1000
        else:
            weekpopularity =float(weekpopularity)
    return weekpopularity
df.weekpopularity=df.weekpopularity.map(lambda weekpopularity:handle(weekpopularity))
```

```python
# 替换sliverticket
def handle(sliverticket):
    if type(sliverticket) == str:
        if sliverticket[-1]=='万':
            sliverticket=float(sliverticket[:-1])*10000
        elif sliverticket[-1]=='千':
            sliverticket=float(sliverticket[:-1])*1000
        else:
            sliverticket =float(sliverticket)
    return sliverticket
df.sliverticket=df.sliverticket.map(lambda sliverticket:handle(sliverticket))
df
```

```python
# 按照周人气排序排序
df.sort_values(by='weekpopularity',inplace=True,ascending=True)
```

```python
# 优化数据表现形式
# 优化chapters
def handle(chapter):
    chapter=chapter[:-2]
    chapter=float(chapter[1:])        
    return chapter
df.chapters=df.chapters.map(lambda chapter:handle(chapter))
df
```

```python
# 优化writer
def handle(writer):
    writer=writer[:-2]        
    return writer
df.writer=df.writer.map(lambda writer:handle(writer))
# df
```

```python
# 处理极端值
# 删除人气小于等于0的
# 删除总字数小于5万的数据
# 删除总章节小于20章的数据
# 删除平均没章节大于20000字的数据
df.drop(df[df.weekpopularity<=0].index,inplace=True)
df.drop(df[df.words<50000].index,inplace=True)
df.drop(df[df.chapters<20].index,inplace=True)
df.drop(df[(df.words/df.chapters)>20000].index,inplace=True)
df
```

```python
df['number']=range(1,898)
```

```python
# 指标计算
# MO列为平均每个字价值银票
# M1列为平均每个章节价值银票
df['M0']=df.sliverticket/df.words
df['M1']=df.sliverticket/df.chapters
df
```

```python
df.drop(df[df.sliverticket>3500000].index,inplace=True)
```

```python
c = (
    Scatter()
    .add_xaxis(df.number)
    .add_yaxis("A", df.sliverticket,symbol_size=6)
    .set_global_opts(
        title_opts=opts.TitleOpts(title="人气对银票的影响"),
        xaxis_opts=opts.AxisOpts(splitline_opts=opts.SplitLineOpts(is_show=False)),
        yaxis_opts=opts.AxisOpts(splitline_opts=opts.SplitLineOpts(is_show=False)),
    
    )
    .set_series_opts(label_opts=opts.LabelOpts(is_show=False))
    .render_notebook()
)
c
```
